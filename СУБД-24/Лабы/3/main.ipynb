```python
import pandas as pd
from pathlib import Path
```

```python
path = Path('../2')
```

```python
dataset = pd.read_csv(str(path / 'last.csv'))
items = pd.read_csv(str(path / 'items.csv'))
people = pd.read_csv(str(path / 'people.csv'))
```

```python
dataset['ID'] = people['Паспорт']
dataset.rename(columns={'ID': 'Паспорт (PK)', 'Инв_ном': 'Инв_ном (FK)'}, inplace=True)
```

```python
items.rename(columns={'Инв_ном': 'Инв_ном (PK)'}, inplace=True)
```

```python
people.rename(columns={'Паспорт': 'Паспорт (PK)'}, inplace=True)
people = people[['Паспорт (PK)', 'ФИО', 'Тел']]
```
