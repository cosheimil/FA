```python
import numpy as np
from scipy.stats import *
from tqdm.notebook import tqdm
from collections import Counter
```

```python
urn1 = ['w1']*9 + ['b1']*13
urn2 = ['w2']*16 + ['b2']*20
n = 10000
A_count = 0
H_count = 0
for k in range(0, n):
    urn3 = list(np.random.choice(urn1, 5, replace=False)) + list(np.random.choice(urn2, 8, replace=False))
    rez = np.random.choice(urn3, 1)
    if rez == 'w1' or rez == 'w2':
        A_count += 1
        if rez == 'w1':
            H_count += 1
print(f'P(A) = {A_count/n}')
print(f'P(H|A) = {H_count/A_count}')
```

```python
sigma_X = 0.5
sigma_Y = 1.1
sigma_Z = 0.4
N = 150_000

lambda_X = sigma_X ** 2
lambda_Y = sigma_Y ** 2
lambda_Z = sigma_Z ** 2

X = poisson(lambda_X)
Y = poisson(lambda_Y)
Z = poisson(lambda_Z)

S = X.rvs(size=N) + Y.rvs(size=N) + Z.rvs(size=N)

print(f"P(S = 3) = {S[S == 3].size / S.size}")
print(f"Наиболее вероятное значение: {Counter(S).most_common()[0][0]}")
print(f"Стандартное отклонение: {S.std()}")
print(f"Ассиметрия: {skew(S)}")

k = []
for _ in range(100):
    S = X.rvs(size=N) + Y.rvs(size=N) + Z.rvs(size=N)
    k.append(kurtosis(S))
k = np.array(k)
print(f"Эксцесс: {k.mean()}")
```

```python
X = rv_discrete(values=[(1, 2, 0), (0.12, 0.21, 0.67)])
N = 10_000

k = []
k_var = []

for _ in tqdm(range(N)):
    i = 0
    t = []

    while not(1 in t and 2 in t):
        value = X.rvs(size=1)[0]
        t.append(value)
        i += 1

    t = ''.join(map(str, t))

    if t.find('2') <= t.find('1'):
        k_var.append(i)

    k.append(i)

k = np.array(k)
print(f"E(X) = {k.mean()}")
print(f"Var(X) = {np.var(k_var, ddof=1)}")
```

```python
a = 23
b = 4

U = uniform(-a, 2 * a)
V = uniform(-b, 2 * b)

count = 0
points = []
print("Генерируем точки")

U_values = U.rvs(size=150_000)
V_values = V.rvs(size=150_000)

points_ = np.vstack([U_values, V_values])
points = points_[:, np.linalg.norm(points_, axis=0).astype(bool)]
print(points)

for i in tqdm(range(N)):
    point1 = np.random.choice(points)
    point2 = np.random.choice(points)
    if np.linalg.norm(point2 - point2) < 9.2:
        count += 1

count/N
```
