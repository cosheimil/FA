```python
n=5
sample = X.rvs(size=n)
mean_values = sample.mean()
print('1) Выборка:', sample)
print('2) Среднее значение суммы выпавших очков:', mean_values)
```

```python
fig,ax =plt.subplots(figsize=(30, 10))
plt.tick_params(labelsize = 40)
ax.xaxis.set_major_locator(ticker.MultipleLocator(200))
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.5))
plt.grid(color='DarkTurquoise', alpha=0.75, linestyle=':', linewidth=3)
plt.ylim(2.5,4.5)
#plt.xlim(-1,3)

summa_mean = []
N=2000
x=[]
for n in range(1,N+1):
    x.append(n+1)
    sample = X.rvs(size=n)
    mean_values = sample.mean()
    summa_mean.append(mean_values)

plt.plot((0, N+1), (3.5, 3.5), color='r',lw=5)
plt.text(-50, 3.45, r'$\frac{7}{2}$',fontsize=50,color='r',weight="bold")
plt.plot(x, summa_mean, color='b',lw=2)
plt.xlabel("Количество подбрасываний (количество испытаний)",size=40)
plt.ylabel("Среднее значение суммы",size=40)
plt.text(600, 4.2, r'Математическое ожидание $\mathbb{E}(X_k)=\frac{7}{2}$',fontsize=40)
plt.text(600, 3.8, r'Среднее значение суммы $\overline{S_n}=\frac{S_n}{n}=\frac{X_1+\ldots X_n}{n}$',fontsize=40)
plt.show()
fig.savefig("LawLargeNumbers_1.pdf", bbox_inches='tight')
```

$$
\begin{array}{l}
X\sim Bin(m;p);\\
\mathbb E(X)=m\cdot p;\\
\mathbb Var(X)=mpq,\qquad\qquad q=1-p.
\end{array}
$$


```python
m=100
p=1/3
X=binom(m,p)
```

```python
n=10
sample=X.rvs(size=n)
mean_values = sample.mean()
print('1) Выборка:', sample)
print('2) Среднее значение суммы:', mean_values)
```

```python
fig,ax =plt.subplots(figsize=(30, 10))
plt.tick_params(labelsize = 40)
ax.xaxis.set_major_locator(ticker.MultipleLocator(200))
ax.yaxis.set_major_locator(ticker.MultipleLocator(3))
plt.grid(color='DarkTurquoise', alpha=0.75, linestyle=':', linewidth=3)
plt.ylim(m*p-5,m*p+5)
#plt.xlim(-1,3)

summa_mean = []
N=2000
x=[]
for n in range(1,N+1):
    x.append(n+1)
    sample=X.rvs(size=n)
    mean_values = sample.mean()
    summa_mean.append(mean_values)

plt.plot((0, N+1), (m*p, m*p), color='r',lw=5)
plt.text(-100, m*p-0.2, r'$\frac{100}{3}$',fontsize=50,color='r',weight="bold")
plt.plot(x, summa_mean, color='b',lw=2)
plt.xlabel("Количество слагаемых",size=40)
plt.ylabel("Среднее значение суммы",size=40)
plt.text(500, m*p+3, r'Математическое ожидание $\mathbb{E}(X_1)=m\cdot p=\frac{100}{3}\approx 33.333$',fontsize=40)
plt.text(500, m*p+1, r'Среднее значение суммы $\overline{S_n}=\frac{S_n}{n}=\frac{X_1+\ldots X_n}{n}$',fontsize=40)
plt.show()
fig.savefig("LawLargeNumbers_2.pdf", bbox_inches='tight')
```

$$
\begin{array}{l}
X\sim Exp(\lambda);\\
\mathbb E(X)=\frac{1}{\lambda};\\
\mathbb Var(X)=\frac{1}{\lambda^2}.
\end{array}
$$

```python
loc=0
lamb=2
scale=1/lamb
X=expon(loc,scale)
```

```python
fig,ax =plt.subplots(figsize=(30, 10))
plt.tick_params(labelsize = 30)
ax.xaxis.set_major_locator(ticker.MultipleLocator(0.5))
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.5))
plt.grid(color='DarkTurquoise', alpha=0.75, linestyle=':', linewidth=2)
#axis('scaled')
plt.ylim(0.,2.5)
plt.xlim(-1,3)


x = np.linspace(-1, 3, 1000)  
y1 = X.pdf(x)  
y2 = X.pdf(x)  
#plt.plot(x, y1, lw=5, alpha=1,color='black')  
#plt.plot(x, y2, 'r--',lw=5, alpha=0.6)  
plt.plot(x, y1, "*", x, y2, "r--") 
plt.show()
```

```python
n=10
sample=X.rvs(size=n)
mean_values = sample.mean()
print('1) Выборка:', sample)
print('2) Среднее значение суммы:', mean_values)
print(r'3) Математическое ожидание E(X) ', X.mean())
```

```python
fig,ax =plt.subplots(figsize=(30, 10))
plt.tick_params(labelsize = 40)
ax.xaxis.set_major_locator(ticker.MultipleLocator(200))
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.25))
plt.grid(color='DarkTurquoise', alpha=0.75, linestyle=':', linewidth=3)
plt.ylim(X.mean()-0.5,X.mean()+0.5)
#plt.xlim(-1,3)

summa_mean = []
N=2000
x=[]
for n in range(1,N+1):
    x.append(n+1)
    sample=X.rvs(size=n)
    mean_values = sample.mean()
    summa_mean.append(mean_values)

plt.plot((0, N+1), (X.mean(), X.mean()), color='r',lw=5)
plt.text(-50, X.mean()-0.02, r'$\frac{1}{2}$',fontsize=50,color='r',weight="bold")
plt.plot(x, summa_mean, color='b',lw=2)
plt.xlabel("Количество слагаемых",size=40)
plt.ylabel("Среднее значение суммы",size=40)
plt.text(500, X.mean()+0.4, r'Математическое ожидание $\mathbb{E}(X_1)=\frac{1}{\lambda}=\frac{1}{2}$',fontsize=40)
plt.text(500, X.mean()+0.2, r'Среднее значение суммы $\overline{S_n}=\frac{S_n}{n}=\frac{X_1+\ldots X_n}{n}$',fontsize=40)
plt.show()
fig.savefig("LawLargeNumbers_3.pdf", bbox_inches='tight')
```
